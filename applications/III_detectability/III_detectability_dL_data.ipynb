# Detectability

```python
def detectability_data(detector, f_low, srate, fres_arr, dPsi, dx_seed):
    filt.f_low = f_low
    filt.srate = srate
    filt.detector = detector
    print(detector)
    MM_detect = 1/(2*sigmasq(filt.tmplt, **filt.kwargs))

    error = 1e-3
    dx_cut = filt.dx_fres_fix_error(error, fres_arr, dPsi)
    
    dx_det, dx_guess = filt.detectable_dx(MM_detect, fres_arr, dPsi, dx_seed, dx_cut)
    joc = filt.read2023_fixed_dL(f=fres_arr)
    
    np.savez_compressed(os.path.join(f'../data/III/dL/{x_type[dPsi]}', f'{x_type[dPsi]}_{detector_name[detector]}.npz'), fres=fres_arr, detect=dx_det, guess=dx_guess, joc=joc, cut=dx_cut)
```
